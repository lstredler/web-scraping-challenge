```python
executable_path = {'executable_path': '/Users/laurenstredler/.wdm/drivers/chromedriver/mac64/90.0.4430.24/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

CREATE CONNECTION <> NASA Mars News
----

```python
#create connection 

mars_url = 'https://mars.nasa.gov/news/'
browser.visit(mars_url) 

time.sleep(1)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
html=browser.html
mars_soup=BeautifulSoup(html,'html.parser')
print(mars_soup)
```

```python
#print title 
title = mars_soup.find('div',class_='content_title').get_text()
print(title)
```

```python
#print paragraph 
paragraph= mars_soup.find('div',class_='article_teaser_body').get_text()
print(paragraph)
```

JPL Mars Space Images - Featured Image
---

```python
#Use splinter to navigate the site and find the image url for the current Featured Mars

search_image_url = "https://www.jpl.nasa.gov/images?search=&category=Mars"
browser.visit(search_image_url) 
html=browser.html
image_soup=BeautifulSoup(html,'html.parser')


featured_image_url=image_soup.select_one('img').get("src")
print(featured_image_url)
```
